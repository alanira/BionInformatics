```python
import matplotlib.pyplot as plt
import numpy as np
species = ['Human', 'Swine', 'Chicken', 'Camel', 'Bat', 'Cat', 'Civet', 'Pangolin', 'Chimpanzee', 'Duck', 'Beluga', 'Goose', 'Falcon']
arr = np.array([957, 785, 309, 265, 181, 57, 5, 5, 4, 3, 2, 2, 1])
s = sum(arr)
print(s)
frequency = arr / s
print(frequency)
xs = range(len(species))
plt.bar(xs, frequency)
plt.xticks(xs, species, fontsize = 12, rotation='vertical')
plt.margins(0.05)
plt.subplots_adjust(bottom=0.15)
plt.xlabel('Species/Hosts', fontsize = 12)
plt.ylabel('Proportion', fontsize = 12)
plt.grid(color = 'green', alpha = 0.2)
plt.title('Host distribution for various coronaviruses', fontsize = 14)
plt.savefig('Host Distribution.png', bbox_inches='tight')
```
